```python
# Remove dollar signs from salary column and convert to float
data_set['original_price'] = data_set['original_price'].str.replace('$','',regex=True)
data_set['original_price'] = data_set['original_price'].astype(float)
```

Now that we have correctly formatted all variables, we need to check for null values (also known as 'NAN' values) in the dataset. Null values can arise due to various reasons such as incomplete data, errors in data entry, or missing values. These null values can have a significant impact on the analysis of the dataset as they can lead to biased results and incorrect conclusions. For example, if we are calculating the average selling price of a product and we do not handle the null values properly, then the average selling price can be skewed towards higher or lower values. Therefore, it is essential to identify and handle null values appropriately before proceeding with the analysis.

As we can see the null values have a big impact on the variable, to the point that we can't even see the values on the graph. How can we deal with these values?


When dealing with missing values, there are several approaches we can take depending on the context of the data. Here are a few common methods:

- Drop the rows with missing values: This is the simplest approach, but it can lead to loss of valuable data if the missing values represent a significant portion of our dataset.

- Impute the missing values: This involves replacing the missing values with a reasonable estimate. Common methods for imputing missing values include replacing with the mean or median value, or using a predictive model to estimate the missing values.

- Ignore the missing values: Depending on the context of our analysis, it may be possible to ignore the missing values without significantly affecting the results.

In the case of the Adidas dataset, since there are only 16 missing values out of a total of 845, it may be reasonable to simply drop these rows using the dropna() function in pandas.

As mentioned above and confirmed by the descriptive analysis of the variables 'selling_price' and 'original_price', these variables have outliers and also we can see in the visualisation of the histogram the two variables are right skewed. One way to deal with these outliers is to logarithmize the data.  This technique can transform a skewed variable into a more normal distribution. 
These techniques can help reduce the impact of outliers on the analysis and improve the accuracy of our predictions.

```python
# Create a list of the variable names to transform
variables_to_transform = ['original_price', 'selling_price']

# Loop through the list of variable names and transform each variable using logarithm
for var in variables_to_transform:
    data_set[f'{var}_log'] = np.log(data_set[var])
```

As we can see from the above graphs, the logarithmic transformation of the data has allowed the outliers to disappear, resulting in a more normalized sample.

By applying the logarithmic transformation, the extreme values are compressed, leading to a more symmetric distribution that resembles a normal distribution. This can facilitate the interpretation of the data and enable the application of more robust statistical techniques.

Reducing outliers and normalizing the data are important to ensure that our analyses are more accurate and representative of the dataset. By eliminating or reducing the influence of extreme values, we can obtain a more accurate view of the trends and patterns present in the data, enabling more reliable analysis and more robust conclusions.


Now let's examine if there are any relationships between the quantitative variables: 'original_price', 'selling_price', 'average_rating', and 'reviews_count'. By analyzing the correlation between these variables, we can gain insights into how they are related to each other.

```python
quantitative_vars = ['original_price', 'selling_price', 'average_rating', 'reviews_count']
corr_matrix = data_set[quantitative_vars].corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Quantitative Variables')
plt.show()
```

As we can see the variables 'original_price' and 'sellingg_price' are strongly correlated, while the remaining variables show little or no correlation. The relationship between 'avarage_rating' and the variables 'original_price' and 'sellingg_price' although it is weak, shows a negative relationship, that is, as the original_price and selling_price increase, the average_rating tends to decrease slightly. However, the correlation coefficient indicates that this relationship is not very strong.

It's important to note that correlation does not imply causation. While the correlation coefficient can provide insights into the relationship between variables, it does not necessarily indicate a causal relationship. Further analysis and domain knowledge are required to understand the underlying factors and determine any causal connections.

Additionally, the lack of significant correlation between variables such as reviews_count and the other variables suggests that they may be independent of each other. This implies that the number of reviews does not have a strong linear relationship with the original_price, selling_price, or average_rating.

Understanding the relationships between variables can help inform decision-making and guide further analysis or modeling strategies.
