Como podemos ver a variavel 'original_price' esta classificada como object, mas na verdade 'e um inteiro, por isso vamos ter que converter

```python
# Remove dollar signs from salary column and convert to float
data_set['original_price'] = data_set['original_price'].str.replace('$','',regex=True)
data_set['original_price'] = data_set['original_price'].astype(float)
```

```python
data_set['crawled_at'] = pd.to_datetime(data_set['crawled_at'])
data_set['Year'] = data_set['crawled_at'].dt.year
data_set['Month'] = data_set['crawled_at'].dt.month
data_set['Day'] = data_set['crawled_at'].dt.day
```

Next, we create a box plot of the 'original_price' column using Matplotlib to identify any outliers. The box plot will show any data points that are more than 1.5 times the interquartile range (IQR) away from the median as individual points, which we can use to identify outliers.

```python
bvars_to_plot = ['original_price', 'selling_price', 'average_rating']

# Plot boxplots of the selected variables
fig, axs = plt.subplots(nrows=1, ncols=len(bvars_to_plot), figsize=(12, 4))
for i, var in enumerate(bvars_to_plot):
    axs[i].boxplot(data_set[var])
    axs[i].set_xlabel(var)
    axs[i].set_ylabel('Value')


plt.tight_layout()
plt.show()
```

```python
# Select the variables for which you want to plot histograms
vars_to_plot = ['original_price', 'selling_price', 'average_rating']

# Plot histograms of the selected variables
fig, axs = plt.subplots(nrows=1, ncols=len(vars_to_plot), figsize=(12, 4))
for i, var in enumerate(vars_to_plot):
    axs[i].hist(data_set[var], bins=20)
    axs[i].set_xlabel(var)
    axs[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
